# Exploratory Data Analysis

## 1. Finding top words

```python
import pickle

with open('../01_LDA_model/pkl_cellar/cv_matrix.pkl', 'rb') as file:
    doc_matrix = pickle.load(file)
    
doc_matrix
```

```python
import pandas as pd
```

```python
# harder to do things across columns than rows

data = doc_matrix.transpose()
data.head()
```

```python
top_dict = {}
for c in data.columns:
    top = data[c].sort_values(ascending=False).head(1)
    top_dict[c] = list(zip(top.index, top.values))
```
